## 1.1 Model The Problem As a Finite Markov Desicion Process:

The solution to this is going to be a slight modification from the midterm since we now have an envirnoment that interacts with the process. Before, if we took an action we always ended up into a specific state. Now once we take an action we can be thrown into another state. This means that the probabilities after we take an action are no longer 1. Let's start off with some quick definitions:

$\mathcal{G} = \{(x,y) \in ([0,m) \times [0,n))$ This is the total m rows and n columns that make up the grid world.
 
$\mathcal{B} \subset \mathcal{G} =$ This variable represents the states that we cannot get into in the grid world. They represent The blacked out squares from the midterm

$\mathcal{T} \subset \mathcal{G} - \mathcal{B}  =$ These are the terminal states and are known as the set of GOAL cells since this is where we are trying to end up with the least cost.

We define an individual state s by it's row and column values (x,y). Next states are denoted by s'.



### State Space:

The state space can be defined by all of the habitable blocks within $\mathcal{G}$. Since $\mathcal{B}$ defines the uninhabitable spaces then we can define the state space as

$\mathcal{S} = \mathcal{G} - \mathcal{B}$

We can further breakdown the state space into terminal and nonterminal states. If we assume that there is no way that the wind can trap us in a block then we can define the terminal states as they are defined in the problem statement. 

$\mathcal{T} \subset \mathcal{S}$

Finally we have the nonterminal states which we define as:

$\mathcal{N} = \mathcal{S} - \mathcal{T}$


### Action Space:

The only actions that we can do are move Up (U), Down (D), Left (L), or Right (R). Thus we define the action space as:

$\mathcal{A} = \{U,D,L,R\}$

### Transition Probability Function:

This is the probability of transitioning from state s to s' depending on the action we choose. This is a little more complicated than last time since actions can now lead to multiple states. Each action now has a probability of going up down or staying in the same spot. and so this needs to be included in the transition probabilities. $p_1(y)$ is the probability of the wind pushing us DOWN by one space if we are in column y. $p_2(y)$ is the probability of the wind pushing us UP by one space if we are in column y. It should also be noted that since we define x to be the row number and y to be the column: **x is the VERTICAL axis and y is the Horizontal axis.**

```python
# Initial Importing:
import sys
sys.path.append("../../RL-book")
from dataclasses import dataclass
from typing import Tuple, Dict, Sequence, Callable
from typing import Iterable, Iterator, Tuple, TypeVar
from rl.markov_decision_process import FiniteMarkovDecisionProcess, MarkovDecisionProcess
from rl.markov_decision_process import FinitePolicy, StateActionMapping, policy_from_q
from rl.markov_process import FiniteMarkovProcess, FiniteMarkovRewardProcess
from rl.distribution import Categorical, Constant, Choose
from scipy.stats import poisson
import numpy as np
from more_itertools import distinct_permutations
import matplotlib.pyplot as plt
from rl.dynamic_programming import value_iteration, policy_iteration, policy_iteration_result
import rl.markov_process as mp
from matplotlib.pyplot import cm
from collections import defaultdict
from rl.function_approx import Tabular
from rl import markov_decision_process
import importlib
importlib.reload(markov_decision_process)
import numpy.linalg as LA

from IPython.core.display import display, HTML
display(HTML("<style>.container{width:80%!important;}"))
```

```python
from typing import Tuple, Sequence, Set, Mapping, Dict, Callable, Optional
from dataclasses import dataclass
from operator import itemgetter
from rl.distribution import Categorical, Choose, Constant
from rl.markov_decision_process import FiniteMarkovDecisionProcess
from rl.markov_decision_process import StateActionMapping
from rl.markov_decision_process import FinitePolicy
from rl.dynamic_programming import value_iteration_result, V

'''
Cell specifies (row, column) coordinate
'''
Cell = Tuple[int, int]
CellSet = Set[Cell]
Move = Tuple[int, int]
'''
WindSpec specifies a random vectical wind for each column.
Each random vertical wind is specified by a (p1, p2) pair
where p1 specifies probability of Downward Wind (could take you
one step lower in row coordinate unless prevented by a block or
boundary) and p2 specifies probability of Upward Wind (could take
you onw step higher in column coordinate unless prevented by a
block or boundary). If one bumps against a block or boundary, one
incurs a bump cost and doesn't move. The remaining probability
1- p1 - p2 corresponds to No Wind.
'''
WindSpec = Sequence[Tuple[float, float]]

possible_moves: Mapping[Move, str] = {
    (-1, 0): 'D',
    (1, 0): 'U',
    (0, -1): 'L',
    (0, 1): 'R'
}


@dataclass(frozen=True)
class WindyGrid:

    rows: int  # number of grid rows
    columns: int  # number of grid columns
    blocks: CellSet  # coordinates of block cells
    terminals: CellSet  # coordinates of goal cells
    wind: WindSpec  # spec of vertical random wind for the columns
    bump_cost: float  # cost of bumping against block or boundary

    def validate_spec(self) -> bool:
        b1 = self.rows >= 2
        b2 = self.columns >= 2
        b3 = all(0 <= r < self.rows and 0 <= c < self.columns
                 for r, c in self.blocks)
        b4 = len(self.terminals) >= 1
        b5 = all(0 <= r < self.rows and 0 <= c < self.columns and
                 (r, c) not in self.blocks for r, c in self.terminals)
        b6 = len(self.wind) == self.columns
        b7 = all(0. <= p1 <= 1. and 0. <= p2 <= 1. and p1 + p2 <= 1.
                 for p1, p2 in self.wind)
        b8 = self.bump_cost > 0.
        return all([b1, b2, b3, b4, b5, b6, b7, b8])

    def print_wind_and_bumps(self) -> None:
        for i, (d, u) in enumerate(self.wind):
            print(f"Column {i:d}: Down Prob = {d:.2f}, Up Prob = {u:.2f}")
        print(f"Bump Cost = {self.bump_cost:.2f}")
        print()

    @staticmethod
    def add_tuples(a: Cell, b: Cell) -> Cell:
        return a[0] + b[0], a[1] + b[1]

    def is_valid_state(self, cell: Cell) -> bool:
        '''
        checks if a cell is a valid state of the MDP
        '''
        return 0 <= cell[0] < self.rows and 0 <= cell[1] < self.columns \
            and cell not in self.blocks

    def get_all_nt_states(self) -> CellSet:
        '''
        returns all the non-terminal states
        '''
        return {(i, j) for i in range(self.rows) for j in range(self.columns)
                if (i, j) not in set.union(self.blocks, self.terminals)}

    def get_actions_and_next_states(self, nt_state: Cell) \
            -> Set[Tuple[Move, Cell]]:
        '''
        given a non-terminal state, returns the set of all possible
        (action, next_state) pairs
        '''
        temp: Set[Tuple[Move, Cell]] = {(a, WindyGrid.add_tuples(nt_state, a))
                                        for a in possible_moves}
        return {(a, s) for a, s in temp if self.is_valid_state(s)}
    
###########################################################################################################

    def get_transition_probabilities(self, nt_state: Cell) \
            -> Mapping[Move, Categorical[Tuple[Cell, float]]]:
        '''
        given a non-terminal state, return a dictionary whose
        keys are the valid actions (moves) from the given state
        and the corresponding values are the associated probabilities
        (following that move) of the (next_state, reward) pairs.
        The probabilities are determined from the wind probabilities
        of the column one is in after the move. Note that if one moves
        to a goal cell (terminal state), then one ends up in that
        goal cell with 100% probability (i.e., no wind exposure in a
        goal cell).
        '''
        d: Dict[Move, Categorical[Tuple[Cell, float]]] = {}
        for a, (r, c) in self.get_actions_and_next_states(nt_state):
            if (r, c) in self.terminals:
                d[a] = Categorical({((r, c), -1.): 1.})
            else:
                ''' write your code here '''
                action_dict = {}
                
                #Step 1: Find the Up and Down Wind Probabilities of the column we are in:
                up_prob = self.wind[c][1]
                down_prob = self.wind[c][0]
                no_wind_prob = 1 - up_prob - down_prob

                
                #Step 2: Wind Does not Blow:
                action_dict[(r, c),-1] =  no_wind_prob
                
                #Step 3: Wind Blows Up
                if self.is_valid_state((r+1,c)):
                    #You get pushed up a cell:
                    action_dict[(r+1, c),-1] =  up_prob
                else:
                    #You get pushed into a wall
                    action_dict[(r, c),-self.bump_cost] =  up_prob
                    
                    
                #Step 4: Wind Blows Down
                if self.is_valid_state((r-1,c)):
                    #You get pushed down a cell:
                    action_dict[(r-1, c),-1] =  down_prob
                else:
                    #You get pushed into a wall
                    action_dict[(r, c),-self.bump_cost] =  down_prob
                    
                    
                #Step 5: Assign action dictionary to action:
                d[a] = Categorical(action_dict)
                
        return d

###########################################################################################################

    def get_finite_mdp(self) -> FiniteMarkovDecisionProcess[Cell, Move]:
        '''
        returns the FiniteMarkovDecision object for this windy grid problem
        '''
        d1: StateActionMapping[Cell, Move] = \
            {s: self.get_transition_probabilities(s) for s in
             self.get_all_nt_states()}
        d2: StateActionMapping[Cell, Move] = {s: None for s in self.terminals}
        return FiniteMarkovDecisionProcess({**d1, **d2})

    def get_vi_vf_and_policy(self) -> Tuple[V[Cell], FinitePolicy[Cell, Move]]:
        '''
        Performs the Value Iteration DP algorithm returning the
        Optimal Value Function (as a V[Cell]) and the Optimal Policy
        (as a FinitePolicy[Cell, Move])
        '''
        return value_iteration_result(self.get_finite_mdp(), gamma=1.)

    @staticmethod
    def epsilon_greedy_action(
        nt_state: Cell,
        q: Mapping[Cell, Mapping[Move, float]],
        epsilon: float
    ) -> Move:
        '''
        given a non-terminal state, a Q-Value Function (in the form of a
        {state: {action: Expected Return}} dictionary) and epislon, return
        an action sampled from the probability distribution implied by an
        epsilon-greedy policy that is derived from the Q-Value Function.
        '''
        action_values: Mapping[Move, float] = q[nt_state]
        greedy_action: Move = max(action_values.items(), key=itemgetter(1))[0]
        return Categorical(
            {a: epsilon / len(action_values) +
             (1 - epsilon if a == greedy_action else 0.)
             for a in action_values}
        ).sample()

    def get_states_actions_dict(self) -> Mapping[Cell, Optional[Set[Move]]]:
        '''
        Returns a dictionary whose keys are the states and the corresponding
        values are the set of actions for the state (if the key is a
        non-terminal state) or is None if the state is a terminal state.
        '''
        d1: Mapping[Cell, Optional[Set[Move]]] = \
            {s: {a for a, _ in self.get_actions_and_next_states(s)}
             for s in self.get_all_nt_states()}
        d2: Mapping[Cell, Optional[Set[Move]]] = \
            {s: None for s in self.terminals}
        return {**d1, **d2}

###########################################################################################################
    
    def get_sarsa_vf_and_policy(
        self,
        states_actions_dict: Mapping[Cell, Optional[Set[Move]]],
        sample_func: Callable[[Cell, Move], Tuple[Cell, float]],
        episodes: int = 10000,
        step_size: float = 0.01
    ) -> Tuple[V[Cell], FinitePolicy[Cell, Move]]:
        '''
        states_actions_dict gives us the set of possible moves from
        a non-block cell.
        sample_func is a function with two inputs: state and action,
        and with output as a sampled pair of (next_state, reward).
        '''
        q: Dict[Cell, Dict[Move, float]] = \
            {s: {a: 0. for a in actions} for s, actions in
             states_actions_dict.items() if actions is not None}
        nt_states: CellSet = {s for s in q}
        uniform_states: Choose[Cell] = Choose(nt_states)
        for episode_num in range(episodes):
            epsilon: float = 1.0 / (episode_num + 1)
            state: Cell = uniform_states.sample()
                
            '''write your code here update the dictionary q initialized above according to the SARSA algorithm's Q-Value Function updates.'''
            
            #Step 1: Start in a random state
            start_state = uniform_states.sample()
            
            #Step 2: Use Epsilon Greedy to determine what action to take
            action = self.epsilon_greedy_action(nt_state = start_state,q=q, epsilon=epsilon)
            
            #Step 3: Determine a new state and reward:
            next_state, reward = sample_func(start_state,action)
            
            #Step 4: Check to see if we ended up in the terminal state:
            if next_state in self.terminals:
                next_action = None
                q[start_state][action] += step_size*(reward  - q[start_state][action] )
            else:
                next_action = self.epsilon_greedy_action(nt_state = next_state,q=q, epsilon=epsilon)
                q[start_state][action] += step_size*(reward + q[next_state][next_action] - q[start_state][action] )
            

        vf_dict: V[Cell] = {s: max(d.values()) for s, d in q.items()}
        policy: FinitePolicy[Cell, Move] = FinitePolicy(
            {s: Constant(max(d.items(), key=itemgetter(1))[0])
             for s, d in q.items()}
        )
        return (vf_dict, policy)
    
###########################################################################################################

###########################################################################################################

    def get_q_learning_vf_and_policy(
        self,
        states_actions_dict: Mapping[Cell, Optional[Set[Move]]],
        sample_func: Callable[[Cell, Move], Tuple[Cell, float]],
        episodes: int = 10000,
        step_size: float = 0.01,
        epsilon: float = 0.1
    ) -> Tuple[V[Cell], FinitePolicy[Cell, Move]]:
        '''
        states_actions_dict gives us the set of possible moves from
        a non-block cell.
        sample_func is a function with two inputs: state and action,
        and with output as a sampled pair of (next_state, reward).
        '''
        q: Dict[Cell, Dict[Move, float]] = \
            {s: {a: 0. for a in actions} for s, actions in
             states_actions_dict.items() if actions is not None}
        nt_states: CellSet = {s for s in q}
        uniform_states: Choose[Cell] = Choose(nt_states)
        for episode_num in range(episodes):
            state: Cell = uniform_states.sample()
            '''
            write your code here
            update the dictionary q initialized above according
            to the Q-learning algorithm's Q-Value Function updates.            '''
            
            #Step 1: Start in a random state
            start_state = uniform_states.sample()
            
            #Step 2: Use Epsilon Greedy to determine what action to take
            action = self.epsilon_greedy_action(nt_state = start_state,q=q, epsilon=epsilon)
            
            #Step 3: Determine a new state and reward:
            next_state, reward = sample_func(start_state,action)
            
            #Step 4: Check to see if we ended up in the terminal state:
            if next_state in self.terminals:
                next_action = None
                q[start_state][action] += step_size*(reward  - q[start_state][action] )
            else:
                #Step 5: Find the action with the maximum Q:
                possible_actions_dict = q[next_state]
                next_action = max(possible_actions_dict,key=possible_actions_dict.get)
                q[start_state][action] += step_size*(reward + q[next_state][next_action] - q[start_state][action] )

        vf_dict: V[Cell] = {s: max(d.values()) for s, d in q.items()}
        policy: FinitePolicy[Cell, Move] = FinitePolicy(
            {s: Constant(max(d.items(), key=itemgetter(1))[0])
             for s, d in q.items()}
        )
        return (vf_dict, policy)
    
###########################################################################################################

    def print_vf_and_policy(
        self,
        vf_dict: V[Cell],
        policy: FinitePolicy[Cell, Move]
    ) -> None:
        display = "%5.2f"
        display1 = "%5d"
        vf_full_dict = {
            **{s: display % -v for s, v in vf_dict.items()},
            **{s: display % 0.0 for s in self.terminals},
            **{s: 'X' * 5 for s in self.blocks}
        }
        print("   " + " ".join([display1 % j for j in range(self.columns)]))
        for i in range(self.rows - 1, -1, -1):
            print("%2d " % i + " ".join(vf_full_dict[(i, j)]
                                        for j in range(self.columns)))
        print()
        pol_full_dict = {
            **{s: possible_moves[policy.act(s).value]
               for s in self.get_all_nt_states()},
            **{s: 'T' for s in self.terminals},
            **{s: 'X' for s in self.blocks}
        }
        print("   " + " ".join(["%2d" % j for j in range(self.columns)]))
        for i in range(self.rows - 1, -1, -1):
            print("%2d  " % i + "  ".join(pol_full_dict[(i, j)]
                                          for j in range(self.columns)))
        print()


if __name__ == '__main__':
    wg = WindyGrid(
        rows=5,
        columns=5,
        blocks={(0, 1), (0, 2), (0, 4), (2, 3), (3, 0), (4, 0)},
        terminals={(3, 4)},
        wind=[(0., 0.9), (0.0, 0.8), (0.7, 0.0), (0.8, 0.0), (0.9, 0.0)],
        bump_cost=4.0
    )
    valid = wg.validate_spec()
    if valid:
        wg.print_wind_and_bumps()
        vi_vf_dict, vi_policy = wg.get_vi_vf_and_policy()
        print("Value Iteration\n")
        wg.print_vf_and_policy(
            vf_dict=vi_vf_dict,
            policy=vi_policy
        )
        mdp: FiniteMarkovDecisionProcess[Cell, Move] = wg.get_finite_mdp()

        def sample_func(state: Cell, action: Move) -> Tuple[Cell, float]:
            return mdp.step(state, action).sample()

        sarsa_vf_dict, sarsa_policy = wg.get_sarsa_vf_and_policy(
            states_actions_dict=wg.get_states_actions_dict(),
            sample_func=sample_func,
            episodes=100000,
            step_size=0.01
        )
        print("SARSA\n")
        wg.print_vf_and_policy(
            vf_dict=sarsa_vf_dict,
            policy=sarsa_policy
        )

        ql_vf_dict, ql_policy = wg.get_q_learning_vf_and_policy(
            states_actions_dict=wg.get_states_actions_dict(),
            sample_func=sample_func,
            episodes=30000,
            step_size=0.01,
            epsilon=0.1
        )
        print("Q-Learning\n")
        wg.print_vf_and_policy(
            vf_dict=ql_vf_dict,
            policy=ql_policy
        )

    else:
        print("Invalid Spec of Windy Grid")
```

I think that I am going to first lay out my understanding of the problem as it is presented. Then I will try to figure out how to solve it. To start the process at t=0 the agent is given a finite amount of a resource. Thus this is probably the only input parameter to get the MDP going. Now at each time step the agent can choose to consue any amount of the resource between 0 and what we had at the end of the last time step. This makes enough sense so far. When we consume a resourse of quantity c of the resource provide a utility of consumption equal to U(c). Let's plot out our CRRA Utility function:

```python
import matplotlib.pyplot as plt
def CRRA(c,gamma):
    utility = (c**(1-gamma))/(1-gamma)
    return utility

for c in range(0,100):
    plt.plot(c,CRRA(c,.5),"r.")
```

Displayed above is the CRRA Utility Function. From what we can see if we decide to consume all 100 in a single step we see that we would get 20. Now if we decided to spend 10 over 10 steps we would get roughly 50 since c = 10 is roughly 5. Obviously we don't want to spend 0 each time step since we don't get any return, but the steepest return is right next to 0. This is where the DISCOUNT parameter $\beta$ comes in. $\beta$ is a number between 0 and 1 that pretty much acts like the gamma parameter from our usual MDPs. Now if $\beta$ is close to 1 it is better to take a smaller amount over a large period of time, but if $\beta$ is close to 0 then it is best to try and get a lot right off the bat since everything will be heavily discounted after that. 


### 2.1 Value Function for a fixed determanistic policy:

In the case of a determanistic policy the value of c that we choose to spend will be a some percentage of the remaining about of resource x. Thus we have an infinite series. In this problem we have a few things going on. We have an exponential decay with disctete integer values of t of our resource given $\theta$ which follows

\begin{equation}
x(t) = x_0 (1-\theta)^t
\end{equation}

We have a consumption value for each of the time steps that follows:

\begin{equation}
c(t) = x_0 \theta (1-\theta)^t
\end{equation}

We also have the utility function that gives us the value of using a certain amount of reward.

\begin{equation}
U(c) = \frac{c^{1-\gamma}}{1-\gamma}
\end{equation}

If we wanted to write out the expected return that we would get it would look something like this.

\begin{equation}
V_{\theta}(x)  = \sum_{t=0}^{\infty} \beta^{t} U(x_0 \theta (1-\theta)^t) = \sum_{t=0}^{\infty} \frac{\beta^{t} \left( (x_0 \theta (1-\theta)^{t})^{1-\gamma} \right) }{1-\gamma}
\end{equation}

However our work is not done as we need to convert this infinite summation into a single equation. What should be noted that if we solve this without the utility function we should be able to represent the value pretty easily. Let's build this up. First if we were to some over just the sold amount we should have it sum to 1 since we expect to get everything back that we sold. We are assuming $x_0 = 1$ for simplicity.

\begin{equation}
\sum_{t=0}^{\infty} \theta (1-\theta)^t = 1
\end{equation}

Now if we add the discount factor $\beta$ we can see that we get a discounted sum where it is better to have a $\beta$ closer to 1. If $\beta = 1$ then we get the previous equation again. 

\begin{equation}
\sum_{t=0}^{\infty} \beta^{t} \theta (1-\theta)^t = \frac{\theta}{\theta \beta - \beta + 1}
\end{equation}

Finally we need to integrate in the effect of the utility function on the infinite sum. This follows the geometric series like the other 2 and we get: 

\begin{equation}
V_{\theta}(x)  = \sum_{t=0}^{\infty} \frac{\beta^{t}(x_0 \theta (1-\theta)^{t})^{1-\gamma} }{1-\gamma} = \sum_{t=0}^{\infty} \frac{\beta^{t} x_0^{1-\gamma} \theta^{1-\gamma} (1-\theta)^{t(1-\gamma)} }{1-\gamma} = \sum_{t=0}^{\infty} \frac{\beta^{t} x_0^{1-\gamma} \theta^{1-\gamma} \left((1-\theta)^t\right)^{1-\gamma}}{1-\gamma} = \sum_{t=0}^{\infty} \frac{\beta^{t} x_0^{1-\gamma} \theta^{1-\gamma} \left((1-\theta)^{1-\gamma}\right)^{t}}{1-\gamma} = \frac{x_0^{1-\gamma} \theta^{1-\gamma}}{1-\gamma} \sum_{t=0}^{\infty} \left(\beta (1-\theta)^{1-\gamma}\right)^{t} = \frac{(x_0 \theta)^{1-\gamma}}{(1-\gamma)\left(1-\beta(1-\theta)^{1-\gamma} \right)}
\end{equation}



## 2.2 Find The Optimal Policy:

To find the optimal policy we will find the policy that maximizes the value function. To do this we can take the derivative of $V_{\theta}(x)$ with repect to $\theta$. We can then set this value equal to 0 and solve for $\theta$ in terms of the remaining variables:

\begin{equation}
\frac{\partial V_{\theta}}{\partial \theta} = \frac{x_0(1-\beta(1-\theta)^{-\gamma}) (\theta x_0)^{- \gamma})}{(\beta (1-\theta)^{1-\gamma}-1 )^2} = 0
\end{equation}

Immediately we can discard a lot of the equation by multiplying it over to the side with zero. We then can solve for $\theta^*$:

\begin{equation}
\frac{x_0(1-\beta(1-\theta)^{-\gamma}) (\theta x_0)^{- \gamma})}{(\beta (1-\theta)^{1-\gamma}-1 )^2} = 0
\end{equation}

\begin{equation}
(1-\beta(1-\theta)^{-\gamma}) = 0
\end{equation}

\begin{equation}
\beta(1-\theta)^{-\gamma} = 1
\end{equation}

\begin{equation}
(1-\theta)^{-\gamma} = \frac{1}{\beta}
\end{equation}

\begin{equation}
\left((1-\theta)^{-\gamma} \right)^{\frac{-1}{\gamma}} = \left(\frac{1}{\beta}\right)^{\frac{-1}{\gamma}}
\end{equation}

\begin{equation}
(1-\theta) = \left(\frac{1}{\beta}\right)^{\frac{-1}{\gamma}}
\end{equation}

\begin{equation}
(1-\theta) = \frac{1}{\beta^{\frac{-1}{\gamma}}}
\end{equation}

\begin{equation}
1-\theta = \beta^{\frac{1}{\gamma}}
\end{equation}

\begin{equation}
\theta^* = 1- \beta^{\frac{1}{\gamma}}
\end{equation}

## 2.3 Optimal Value Function:

Let's start off with the value function that we solved for in the first part:

\begin{equation}
V_{\theta}(x)   = \frac{(x_0 \theta)^{1-\gamma}}{(1-\gamma)\left(1-\beta(1-\theta)^{1-\gamma} \right)}
\end{equation}

Now we can replace the policy with the optimal policy:

\begin{equation}
V^*(x)   = \frac{(x_0 (1- \beta^{\frac{1}{\gamma}}))^{1-\gamma}}{(1-\gamma)\left(1-\beta(1-(1- \beta^{\frac{1}{\gamma}}))^{1-\gamma} \right)}
\end{equation}

## 3.1 Model as an MDP: 

When it comes to modelling this as an MDP...

### State Space:

First, I want to determine all of the parameters that we need to determine a state. I think that we need remaining loan amount $B_i$, Mortgage rate (M), and remaining months of payments left (m). The rest of the useful metric like monthly pricnciple and interest can be calculated from these values following the formulas that were presented in the question. 

$\mathcal{S} = \{(L,M,B_i,m) : L > 0, M \geq 0, B_i \geq 0, m \geq 0 \}$

The terminal state for this process is when we have fully paid off our loan. And either B_i = 0 or m = 0 which signifies that we have made it to the end of our payments. These things happen at the same time. It should also be noted that we don't actually need L as B_i captures all of the information that it would give. I am just keeping it in the state as bookkeeping. 

$\mathcal{T} \subset \mathcal{S} = \{(L,M,B_i,m) : L > 0, M \geq 0, B_i = 0, m = 0  \}$

$\mathcal{N} = \mathcal{S} - \mathcal{T}$

### Action Space:

The action space is pretty easy to define as there are only 2 actions that we can do which are: Choose to ignore the bank's offer of the updated mortgage rate and continue on with the payments for your current mortgage ($\mathcal{I}$), or you can choose to cancel ($\mathcal{C}$) your current mortgage and enter into a new mortgage where the loan amount will be equal to the loan balance on your current mortgage and the mortgage rate will be the updated mortgage rate the bank just offered you.

$\mathcal{A} = \{\mathcal{I}, \mathcal{C} \}$

### Reward Space:

To keep track of the rewards I think that it makes sense to make the reward equal to the negative payments that we make. as well as any cost that we need to pay. To clarify what everything means, L' and M' represent the new loan and morgage rate. L' is equal to the previous balance, but I think it's important to include it as L'. The i counter represents the timestep since the start and thus it is always increasing and does not reset with a new loan, the m variable keeps track of how many months of payment we have left. This notation carries over into the transitions.

\begin{gather}
\mathcal{R}_T(s,a,r,s') = \left\{
\begin{array}{ll}
r= -P_i - I_i & a = \mathcal{I} & s'(L=L, & M=M, & B_{i+1} = B_{i} - P_{i} & m=m-1 ), & s' \in \mathcal{S}\\
r= -P_i - I_i - C & a = \mathcal{C} & s'(L=L' & M = M', & B_{i+1} = L' & m = n), & s' \in \mathcal{S}\\
0 & \textrm{otherwise}
\end{array}
\right\}
\end{gather}

### Transitions:

In this example, there really is not a level of randomness in the environment. Obviously we don't know what the new mortgage rate is going to be, but that only effects the action that we take. In otherwords we know that the action we take leads us to a new state with a probability of 1 as long as that new state is in the possible states.

\begin{gather}
\mathcal{P}(s,a,s') = \left\{
\begin{array}{ll}
1 & a = \mathcal{I} & s'(L=L, & M=M, & B_{i+1} = B_{i} - P_{i} & m=m-1 ), & s' \in \mathcal{s}\\
1 & a = \mathcal{C} & s'(L=L' & M = M', & B_{i+1} = L' & m = n), & s' \in \mathcal{s}\\
0 & \textrm{otherwise}
\end{array}
\right\}
\end{gather}

### Discount Factor:

The discount factor is going to be related to the time value of money. I think this part is the most confusing to me so I am going to try and lay it out well. since we know that we evaluate the rewards monthly and the rate at which the time value of money increases is on a yearly time scale the 1/12 allows us to convert between them. Then when we write out the expected return: 

$G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \dots$
