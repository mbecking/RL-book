```python
def generate_traces_as_single_steps(FiniteMDP, policy, start_state_dist, num_steps):
    counter = 0
    traces = []
    
    FiniteMRP = FiniteMDP.apply_finite_policy(policy)
    
    for i, episode in enumerate(FiniteMRP.reward_traces(start_state_dist)):

        if counter >= num_steps:
            break

        for step in episode:
            counter += 1
            traces.append(step)
    
    return traces


    
    
```

```python
def get_feature_vec(state,feature_functions) -> np.ndarray:
        return np.array( [f(state) for f in feature_functions])
```

```python
X = TypeVar('X')
S = TypeVar('S')
def least_squares_temporal_difference(feature_functions: Sequence[Callable[[X], float]],
                                      traces: Iterable[mp.TransitionStep[S]],
                                      gamma : float) -> Sequence[float]:
    """Computes the value function of an MRP using the LSTD method.
    
    Parameters:
        feature_functions : 
        traces
        gamma:"""
    
    n = len(feature_functions)
    
    #Step 1: Intial A and b Matricies:
    A = np.zeros((n,n))
    b = np.zeros((1,n))
            
    #Step 3: Fill out what the x_values that go into the feature functions will be?
    
    #Step 2: Step through all of the single transitions and update A and b
    for i,transtion in enumerate(traces):
        phi_1 = get_feature_vec(transtion.state,feature_functions)
        phi_2 = get_feature_vec(transtion.next_state,feature_functions)
        
        
        
        A += np.outer(phi_1,(phi_1 - gamma * phi_2))
        b += phi_1 * transtion.reward
        
    print (A)
    print (b)
    
    weights = np.dot(LA.inv(A),b.T)
    

    return weights
```

```python
#Create Class:
@dataclass(frozen=True)
class Village_State:
    num_villagers_alive: int
    Vampire_is_alive : bool
        
VampireMapping = StateActionMapping[Village_State, int]

class Vampire_MDP_Finite(FiniteMarkovDecisionProcess[Village_State, int]):
    def __init__(self,num_villagers_alive:int):
        self.num_villagers_alive = num_villagers_alive
        super().__init__(self.get_action_transition_reward_map())
    
    def get_action_transition_reward_map(self) -> VampireMapping:
        #d: Dict[Village_State, Dict[str, Categorical[Tuple[Village_State,float]]]] = {}
        d: Dict[Village_State, Dict[int, Categorical[Tuple[Village_State,float]]]] = {}

        for villagers in range(1,self.num_villagers_alive+1):
            START_STATE = Village_State(num_villagers_alive = villagers,Vampire_is_alive=True)
            action_dict = {}
            
            for poison_amount in range(villagers+1):
                #action_str = "Poison " + str(poison_amount) + " villigers"
                action_str = (poison_amount) 
                END_STATE_Murdered = Village_State(num_villagers_alive = villagers-poison_amount,Vampire_is_alive=False)
                END_STATE_Survived = Village_State(num_villagers_alive = villagers-poison_amount-1,Vampire_is_alive=True)
                
                prob_dict = {}
                
                vampire_murdered_prob = poison_amount/villagers
                vampire_survive_prob = 1 - poison_amount/villagers
                
                Reward_vampire_survives = 0
                Reward_vampire_is_murdered = villagers - poison_amount
                
                prob_dict[(END_STATE_Murdered,Reward_vampire_is_murdered)] = vampire_murdered_prob
                prob_dict[(END_STATE_Survived,Reward_vampire_survives)] = vampire_survive_prob
                
                action_dict[action_str] =Categorical(prob_dict)
                
                d[END_STATE_Murdered] = None
            d[Village_State(num_villagers_alive = 0,Vampire_is_alive=True)]=None
            d[Village_State(num_villagers_alive = 0,Vampire_is_alive=False)]=None
            d[START_STATE] = action_dict
        return d
    
Vampire_MDP = Vampire_MDP_Finite(num_villagers_alive=10)
```

```python
def random_policy(FiniteMDP):
    policy_dict = {}
    for state in FiniteMDP.states():
        actions = list(FiniteMDP.actions(state))
        n = len(actions)
        if n > 0:
            uniform_actions = Categorical({action:1/n for action in actions})
            random_action = uniform_actions.sample()
            policy_dict[state] = Constant(random_action)
        else:
            policy_dict[state] = None
            
    return FinitePolicy(policy_dict)
```

# Problem 2: Implment LSPI Algorithm in Python:

Firstly, LSPI stands for Least Squares Policy Iteration. It is a method to calculate the optimal policy and optimal value function. It is in the **Control** domain. Iy uses the least squares fitting to determine the a policy and then uses that to improve. To develop this algorithm we will be taking the approach that is laid out in the David Silver slides:



I think that D needs to have randomly choosen actions in the traces.

The logic in the code is pretty straight forward, but I want to break into into steps anyways:

1. Inputs: FiniteMDP, initial_policy, Batch_Traces, feature_functions, gamma)
2. Use the initial_policy to create an MRP
3. Evaluate the MRP using LSTDQ to determine Q for all states and actions.
4. For all states act greedily for Q to determine a new better policy.
5. Repeat this process until the policy stops changing
6. Outputs: optimal_policy, Q_optimal

```python
def get_Q_feature_vec(feature_functions,state,action):
    return np.array( [f(state,action) for f in feature_functions])


def f1(Village_State,action):
    return Village_State.num_villagers_alive

def f2(Village_State,action):
    return int(Village_State.Vampire_is_alive)

def f3(Village_State,action):
    if action is not None:
        return Village_State.num_villagers_alive * action
    else:
        return 0

def f4(Village_State,action):
    if action is not None:
        return action
    else:
        return 0
    
def f5(Village_State,action):
    if action is not None:
        return Village_State.num_villagers_alive*(Village_State.num_villagers_alive - action)
    else:
        return 0
    
feature_functions = [f1,f2]
```

```python
def generate_universal_policy(FiniteMDP):
    """Creates a nondetermanistic policy that tests all actions for a state equally"""
    
    policy_dict = {}
    for state in FiniteMDP.states():
        actions = list(FiniteMDP.actions(state))
        n = len(actions)
        if n > 0:
            uniform_actions = Categorical({action:1/n for action in actions})
            #random_action = uniform_actions.sample()
            policy_dict[state] = uniform_actions
        else:
            policy_dict[state] = None
            
    return FinitePolicy(policy_dict)
```

```python
def generate_traces_with_random_actions(FiniteMDP,start_distribution,universal_policy,num_steps):
    traces = []
    counter = 0
    for i, episode in enumerate(FiniteMDP.action_traces(start_distribution,universal_policy)):

        counter += 1
        
        if counter >= num_steps:
            break

        for step in episode:
            counter += 1
            traces.append(step)
    
    return traces

universal_policy = generate_universal_policy(Vampire_MDP)
start_state_distribution = Categorical({Village_State(10,True):.5,Village_State(9,True):.5})
traces = generate_traces_with_random_actions(Vampire_MDP, start_state_distribution, universal_policy, 10000)

# for t in traces:
#     print (t)
```

```python
def Calculate_Q_approx(feature_functions, weights, state, action):
    
    feature_vec = get_Q_feature_vec(feature_functions,state,action)
    Q = np.matmul(feature_vec,weights)
    
    return Q
```

```python
def get_all_actions_for_state(State):
    #For vampire Problem
    actions = []
    for i in range(State.num_villagers_alive+1):
        actions.append(i)
    if State.Vampire_is_alive ==False:
        actions = []
    return actions   

get_all_actions_for_state(Village_State(10,True))
```

```python
X = TypeVar('X')
S = TypeVar('S')
def LSTDQ(feature_functions: Sequence[Callable[[X], float]],
          traces: Iterable[mp.TransitionStep[S]],
          gamma : float) -> Sequence[float]:
    
    """Computes the action value function of an MDP using the LSTDQ method.
    
    Parameters:
        feature_functions : 
        traces
        gamma:"""
    
    n = len(feature_functions)
    
    #Step 1: Intial A and b Matricies:
    A = np.zeros((n,n))
    b = np.zeros((1,n))
    
    weights = np.zeros((n,1))
            
    #Step 3: Fill out what the x_values that go into the feature functions will be?
    
    #Step 2: Step through all of the single transitions and update A and b
    for i,transtion in enumerate(traces):
        
        
        #Vampire problem specific part:
        actions = get_all_actions_for_state(transtion.next_state)
        Q_max = -1
        next_action = None
        for action in actions:
            #Evaluate Q of being in that next state and choosing that action:
            Q = Calculate_Q_approx(feature_functions, weights, transtion.next_state, action)
            if Q > Q_max:
                Q_max = Q
                next_action = action
        
        
        phi_1 = get_Q_feature_vec(feature_functions,transtion.state,transtion.action)
        phi_2 = get_Q_feature_vec(feature_functions,transtion.next_state,next_action)
        
        
    A += np.outer(phi_1,(phi_1 - gamma * phi_2))
    b += phi_1 * transtion.reward

    print (A)
    print (b)

    weights = np.dot(LA.inv(A),b.T)
    

    return weights
```
