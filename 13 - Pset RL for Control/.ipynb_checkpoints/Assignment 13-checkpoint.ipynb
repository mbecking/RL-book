```python
def random_policy(FiniteMDP):
    policy_dict = {}
    for state in FiniteMDP.states():
        actions = list(FiniteMDP.actions(state))
        n = len(actions)
        if n > 0:
            uniform_actions = Categorical({action:1/n for action in actions})
            random_action = uniform_actions.sample()
            policy_dict[state] = Constant(random_action)
        else:
            policy_dict[state] = None
            
    return FinitePolicy(policy_dict)
```

```python
def get_1_episode_of_MRP(FiniteMDP,policy,start_state_distribution):
    FiniteMRP = FiniteMDP.apply_finite_policy(policy)
    episode = list(FiniteMRP.simulate_reward(start_state_distribution))
    
    return episode
```

# Problem 1: Tabular Monte Carlo - GLIE:

General approch for any MDP:

1. Create MDP as a finiteMDP Class
2. We are create a function to return an optimal value function and optimal policy. 
3. Inputs: (MDP,epsilon)
4. Outputs: (Optimal Value Function, Optimal Policy)
5. Steps in the function:

```python
def tabular_monte_carlo_GLIE(FiniteMDP,gamma,start_state_distribution):
    
    #Step 1: Create all dictionaries:
    Q_approx   : Dict[Tuple(state,action),float] = defaultdict(def_val)
    num_visits : Dict[Tuple(state,action),int] = defaultdict(def_val)
    
    #Step 2: Build a random policy from the MDP:
    policy = random_policy(FiniteMDP)
    
    #Step 3: begin while loop to solve for the optimal policy:
    counter_k = 0
    tol = .001
    SSE = 1.0
    
    while SSE > tol and counter_k < 100000:
        counter_k += 1
        
        #Step 4: get a single episode for the monte-carlo simulation to follow
        episode = get_1_episode_of_MRP(FiniteMDP,policy,start_state_distribution)
        
        #Step 5: Loop through all of the steps in the episode:
        for n,transition in enumerate(episode): #SARS

            #Step 6: Calculate the reward from this state to the end
            actual_reward = 0
            
            #Step 7:
            for i,transition_2 in enumerate(episode): #(All of the remaining steps and their rewards)
                if i >= n:

                    actual_reward += gamma*transition_2.reward
                    
            #Step 8: pull the action from the policy that we are using and update visit number and Q_approx
            action = policy.act(transition.state)
            
            action = action.value
            num_visits[(transition.state,action)] += 1
            Q_approx[(transition.state,action)] += 1/(num_visits[(transition.state,action)]) * (actual_reward - Q_approx[(transition.state,action)])
            
            
        #Step 9: Find the Sum of Square error between this and the previous value.
        SSE = 1 #Maybe I will implement (idk tho)
        
        #Step 10: Update the policy using epsilon Greedy
        Qpolicy = markov_decision_process.policy_from_q(q=Tabular(Q_approx), mdp=FiniteMDP, ϵ=1/(min(num_visits.values())+1))
        
        
        policy_dict = {}
        for state in FiniteMDP.states():
            action_dist = Qpolicy.act(state)
            
            if action_dist is not None:
                if type(action_dist) == Choose:
                    policy_dict[state] = Constant(action_dist.sample())
                else:
                    policy_dict[state] = action_dist
            else:
                policy_dict[state] = None
        
        policy = FinitePolicy(policy_dict)
        
    Qpolicy = markov_decision_process.policy_from_q(q=Tabular(Q_approx), mdp=FiniteMDP, ϵ=0)

    policy_dict = {}
    for state in FiniteMDP.states():
        action_dist = Qpolicy.act(state)

        if action_dist is not None:
            if type(action_dist) == Choose:
                policy_dict[state] = Constant(action_dist.sample())
            else:
                policy_dict[state] = action_dist
        else:
            policy_dict[state] = None
        
    policy = FinitePolicy(policy_dict)
    
    return policy,Q_approx,num_visits
            
```

## Test Code: (Vampire Problem)

```python
#Create Class:
@dataclass(frozen=True)
class Village_State:
    num_villagers_alive: int
    Vampire_is_alive : bool
        
VampireMapping = StateActionMapping[Village_State, int]

class Vampire_MDP_Finite(FiniteMarkovDecisionProcess[Village_State, int]):
    def __init__(self,num_villagers_alive:int):
        self.num_villagers_alive = num_villagers_alive
        super().__init__(self.get_action_transition_reward_map())
    
    def get_action_transition_reward_map(self) -> VampireMapping:
        d: Dict[Village_State, Dict[str, Categorical[Tuple[Village_State,float]]]] = {}

        for villagers in range(1,self.num_villagers_alive+1):
            START_STATE = Village_State(num_villagers_alive = villagers,Vampire_is_alive=True)
            action_dict = {}
            
            for poison_amount in range(villagers+1):
                action_str = "Poison " + str(poison_amount) + " villigers"
                END_STATE_Murdered = Village_State(num_villagers_alive = villagers-poison_amount,Vampire_is_alive=False)
                END_STATE_Survived = Village_State(num_villagers_alive = villagers-poison_amount-1,Vampire_is_alive=True)
                
                prob_dict = {}
                
                vampire_murdered_prob = poison_amount/villagers
                vampire_survive_prob = 1 - poison_amount/villagers
                
                Reward_vampire_survives = 0
                Reward_vampire_is_murdered = villagers - poison_amount
                
                prob_dict[(END_STATE_Murdered,Reward_vampire_is_murdered)] = vampire_murdered_prob
                prob_dict[(END_STATE_Survived,Reward_vampire_survives)] = vampire_survive_prob
                
                action_dict[action_str] =Categorical(prob_dict)
                
                d[END_STATE_Murdered] = None
            d[Village_State(num_villagers_alive = 0,Vampire_is_alive=True)]=None
            d[Village_State(num_villagers_alive = 0,Vampire_is_alive=False)]=None
            d[START_STATE] = action_dict
        return d
    
Vampire_MDP = Vampire_MDP_Finite(num_villagers_alive=10)
```

```python
start_state_distribution = Categorical({Village_State(10,True):.5,Village_State(9,True):.5})
optimal_policy,Q_approx,num_visits = tabular_monte_carlo_GLIE(Vampire_MDP,1,start_state_distribution)

print ("Optimal Policy")
print (optimal_policy)
print ()

print ("Approximate State-Action Value Function (Q):")
for Q in Q_approx:
    print (Q,"  ",Q_approx[Q])
```

# Problem 2: Tabular S.A.R.S.A

```python
def tabular_SARSA(FiniteMDP,gamma,start_state_distribution):
    #Step 1: Create all dictionaries:
    Q_approx   : Dict[Tuple(state,action),float] = defaultdict(def_val)
    num_visits : Dict[Tuple(state,action),int] = defaultdict(def_val)
    
    
    #Step 2: Build a random policy from the MDP:
    policy = random_policy(FiniteMDP)
    
    #Step 3: begin while loop to solve for the optimal policy:
    counter_k = 0
    tol = .001
    SSE = 1.0
    
    while SSE > tol and counter_k < 100000:
        counter_k += 1
        
        #Step 4: get a single episode for the monte-carlo simulation to follow
        episode = get_1_episode_of_MRP(FiniteMDP,policy,start_state_distribution)
        
        #Step 5: Loop through all of the steps in the episode:
        for n,transition in enumerate(episode): #SARS

            #Step 6: First Action
            action = policy.act(transition.state).value
            
            #Step 7: Next Action
            if policy.act(transition.next_state) is None:
                next_action = policy.act(transition.next_state)
            else:
                next_action = policy.act(transition.next_state).value
            
            #Step 8: Update Num of visits and Q approx
            num_visits[(transition.state,action)] += 1
            Q_approx[(transition.state,action)] += 1/(num_visits[(transition.state,action)]) * (transition.reward + gamma*Q_approx[(transition.next_state,next_action)] - Q_approx[(transition.state,action)])
            
            
        #Step 9: Find the Sum of Square error between this and the previous value.
        SSE = 1 #Maybe I will implement (idk tho)
        
        #Step 10: Update the policy using epsilon Greedy
        Qpolicy = markov_decision_process.policy_from_q(q=Tabular(Q_approx), mdp=FiniteMDP, ϵ=1/(min(num_visits.values())+1))
        
        
        policy_dict = {}
        for state in FiniteMDP.states():
            action_dist = Qpolicy.act(state)
            
            if action_dist is not None:
                if type(action_dist) == Choose:
                    policy_dict[state] = Constant(action_dist.sample())
                else:
                    policy_dict[state] = action_dist
            else:
                policy_dict[state] = None
        
        policy = FinitePolicy(policy_dict)
        
    Qpolicy = markov_decision_process.policy_from_q(q=Tabular(Q_approx), mdp=FiniteMDP, ϵ=0)

    policy_dict = {}
    for state in FiniteMDP.states():
        action_dist = Qpolicy.act(state)

        if action_dist is not None:
            if type(action_dist) == Choose:
                policy_dict[state] = Constant(action_dist.sample())
            else:
                policy_dict[state] = action_dist
        else:
            policy_dict[state] = None
        
    policy = FinitePolicy(policy_dict)
    
    return policy,Q_approx,num_visits
```
